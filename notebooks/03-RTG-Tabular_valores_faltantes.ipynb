## Variables / columnas

### Resumenes por variable

```python
riskfactors.missing.variable_summary()
```

Es interesante, por ejemplo, que la variable Pregnant tiene un 87% de valores faltantes.

```python
riskfactors_nas = riskfactors.missing.na_summary()
riskfactors_nas.sort_values(by=['n_variables'], ascending=False)
```

```python

sns.barplot(
    data = riskfactors_nas,
    x="n_na",
    y="n_variables"
)
```

Debemos empezar a sospechar que las variables faltantes tengan relaciones entre sí. Por ejemplo que los valores faltantes en un lugar también falten en otro.
