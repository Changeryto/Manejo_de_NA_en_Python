## Exploraciones visuales b√°sicas

### El uso de porcentaje de faltantes por variable

```python
riskfactors_na_vars = riskfactors.missing.variable_summary().sort_values(by="n_na", ascending=False)

sns.barplot(
    data = riskfactors_na_vars,
    y = "variable",
    x = "na_percentage"
)
plt.xlabel("% de valores faltantes")
plt.ylabel("Variable")
```

Eventualmente una tabla se vuelve insuficiente

## Observar valores que conjuntamente aparecen faltantes.

```python
sns.heatmap(
    riskfactors.isna()
)

```

Puede observarse que pregnant tiene muchos valores faltantes, adem√°s debe observarse que hay valores cuyo contexto est√° muy relacionado, y cuando una no aplica, el resto tampoco (como el smoke o el drink).

```python
riskfactors.missing.row_summary()[["n_na"]].value_counts().to_frame().reset_index()
```

```python
riskfactors.missing.row_summary()[["n_na"]].value_counts().to_frame().reset_index().pipe(
    lambda data : sns.barplot(data, y="count", x="n_na")
)
plt.xlabel("N√∫mero de NA en el registro")
plt.ylabel("N√∫mero de casos")
```

```python
riskfactors.missing.row_summary(50)[["prc_na"]].reset_index()
```

```python
span = 10

riskfactors.missing.row_summary(span)[["prc_na"]].reset_index().pipe(
    lambda data : sns.barplot(data, x="index", y="prc_na")
)
plt.xlabel(f"Intervalo de {span}")
plt.ylabel("% de NA")
plt.ylim([0, 80])
```

Es √∫til observar si hay intervalos particularmente altos en valores faltantes.

En ocasiones ya hay librer√≠as que ayudan a manejar valores faltanes.
